2.4   Encode the sentiment column values as 1 or 0

```python
df['Sentiment']=df['Sentiment'].replace({'negative':'0','positive':'1'})
df.head(5)
```

3.  Data preprocessing

3.1 Remove special characters and html tags

```python
from bs4 import BeautifulSoup
import re
```

```python
def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()
```

```python
def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
df['Review'] = df['Review'].apply(remove_html_tags)
df['Review'] = df['Review'].apply(remove_special_chars)
df.head(5)
```

###  INFERENCE:   <font color='brown'> All special characters, numbers and html tags are removed from the dataset.</font>

3.2 - Convert reviews into lowercase

```python
df['Review'] = df['Review'].str.lower()
df.head(5)
```

###  INFERENCE:   <font color='brown'> Entire text data is converted to lowercase.</font>

3.3 - Removal of stop words

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stp_wds = stopwords.words('english')
stp_wds.remove('not')
print(stp_wds)
```

```python
def remove_stopwords(text):
    words = text.split()  # Split the text into words
    filtered_words = [i for i in words if i not in set(stp_wds)]
    return ' '.join(filtered_words)  # Join the filtered words back into a sentence
```

```python
df['Review'] = df['Review'].apply(remove_stopwords)
df.head(5)
```

###  INFERENCE:   <font color='brown'> Stop words like articles, pronouns, common verb, prepositions, conjunctions are removed.</font>

3.4 - Apply stemming

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem_text(text):
    words = text.split()  # Tokenize the text into words
    stemmed_words = [ps.stem(i) for i in words]  # Stem each word
    return ' '.join(stemmed_words)  # Join the stemmed words back into a sentence
```

```python
df['Review'] = df['Review'].apply(stem_text)
df.head(10)
```

###  INFERENCE:   <font color='brown'> Stemming is the process of removing inflections from the word and getting the base or root word.</font>

4.  Convert review text into feature vector and classification

4.1 - create BOW model 

```python
corpus = []
for i in range(0, len(df)):
    corpus.append(df['Review'][i])
corpus    
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
X = cv.fit_transform(corpus).toarray()
X
```

###  INFERENCE:   <font color='brown'>"Bag of Words" (BoW) is a common technique in natural language processing (NLP) and text analysis. It converts a collection of text documents into numerical features that can be used to give inputs into an ML model.</font>

4.2 - Training data and Test data -split

```python
samp=df.iloc[:]
samp.head(5)
```

```python
Y= df.iloc[:,0]
Y.head(5)
```

```python
Y=df.iloc[:,0].values
Y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 40)
```

4.3 - Define the Classification model (any one like NB, SVM or Random forest) train the model

```python
# Training model using Naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(X_train, y_train)
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

4.4  Classification metrics analysis and plot the confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion_matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
print("Accuracy_score:",accuracy_score(y_test, y_pred))
```

###  INFERENCE:   <font color='brown'>The accuracy score of the ML model constructed by using Multinomial Naive Bayes classifier and CountVectorizer is 0.84 </font>

4.5 - predict the class for your own review

```python
test_review1 = "In the last quarter of 2010, Espresso financials witnessed a remarkable performance."
tst_review_bow1 = cv.transform([test_review1])
tst_pred1 = classifier.predict(tst_review_bow1)

if tst_pred1 == ['1']:
    print("Predicted class of test review is :",tst_pred1, "--> POSITIVE")
else:
    print("Predicted class of test review is :",tst_pred1, "--> NEGATIVE")
```

```python
test_review2 = "Brownwich Newyork recently laid off several employees at its 4 sites in the city"
tst_review_bow2 = cv.transform([test_review2])
tst_pred2 = classifier.predict(tst_review_bow2)

if tst_pred2 == ['1']:
    print("Predicted class of test review is :",tst_pred2, "--> POSITIVE")
else:
    print("Predicted class of test review is :",tst_pred2, "--> NEGATIVE")
```

4.6 - create feature vector using tf-idf

4.7 Define the Classification model (any one like NB, SVM or Random forest) train the model

```python
from sklearn.metrics import classification_report
report_tf = classification_report(ytf_test, ytf_pred)
print(report_tf)
```

```python
print("Accuracy_score:",accuracy_score(ytf_test, ytf_pred))
```
